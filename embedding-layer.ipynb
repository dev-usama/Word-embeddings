## Data Preprocessing

```python
dataset = pd.read_csv("/kaggle/input/movie-reviews/IMDB Dataset.csv")
```

```python
dataset.head(10)
```

```python
labelEncoder = LabelEncoder()
```

```python
labels = dataset["sentiment"]
labels = labelEncoder.fit_transform(labels)
```

```python

# Calculate the length of each string in the 'Text' column
dataset['Length'] = dataset['review'].apply(" ".join)
dataset['Length'] = dataset['review'].apply(len)

# Find the maximum length
max_length = dataset['Length'].max()
```

```python
tokenizer = Tokenizer()

# Fit the tokenizer on the text
tokenizer.fit_on_texts(dataset['review'])

# Get the word index (vocabulary)
word_index = tokenizer.word_index
```

```python
vocab_size = len(word_index) # word_index is our vocabulary which is of type dictionary
```

```python
encoded_reviews = [one_hot(x, vocab_size) for x in dataset["review"]]
```

```python
feature_size = 100
padded_reviews = pad_sequences(encoded_reviews, maxlen=max_length, padding="post")
```

```python
padded_reviews
```

```python
model = tf.keras.Sequential(
    [
        Embedding(vocab_size, feature_size, input_length=max_length, name="embedding"),
        Flatten(),
        layers.Dense(units=1, activation="sigmoid")
    ]
)
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model.fit(padded_reviews, labels, epochs=2)
```

```python
model.save("word_embeddings_2.keras")
```
